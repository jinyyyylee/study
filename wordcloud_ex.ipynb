```
#wordcloud 생성
from wordcloud import WordCloud
from konlpy.tag import Twitter 
from collections import Counter
#matplotlib.pyplot: 시각화하는 대표적인 패키지
import matplotlib.pyplot as plt
#pandas : 파일을 읽어옴
import pandas as pd
```

```
df = pd.read_csv(r"/content/drive/My Drive/Colab Notebooks/Youtube04-Eminem.csv")
```

```
#pandas의 데이터프레임(df)는 아주아주 중요합니다
print(df.head(0))
```

```
print(df.CONTENT)
```

```
corpus=''
for val in df.CONTENT:#val의 타입 : Serial! pandas를 구성하는
  val = str(val)#Serial ->> str타입으로 형 변환
  tokens = val.split()#공백문자를 기준으로 쪼개기->list타입으로 변환

  for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()#소문자로 변경
  #wordcloud로 표현할 때에는 소문자 대문자를 구별하기 때문에 모두 소문자로 변환
  for words in tokens:
    corpus = corpus + words + ' '#단어를 결합

print(corpus)
```

```
wordcloud = WordCloud().generate(corpus)

plt.figure(figsize = (8,8))#figure : 그림의 사이즈
plt.imshow(wordcloud)#imshow : 이미지를 보여줌
plt.axis("off")#가로축 세로축 눈금 x
plt.show()
```
