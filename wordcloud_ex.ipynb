```
#wordcloud ÏÉùÏÑ±
from wordcloud import WordCloud
from konlpy.tag import Twitter 
from collections import Counter
#matplotlib.pyplot: ÏãúÍ∞ÅÌôîÌïòÎäî ÎåÄÌëúÏ†ÅÏù∏ Ìå®ÌÇ§ÏßÄ
import matplotlib.pyplot as plt
#pandas : ÌååÏùºÏùÑ ÏùΩÏñ¥Ïò¥
import pandas as pd
```

```
df = pd.read_csv(r"/content/drive/My Drive/Colab Notebooks/Youtube04-Eminem.csv")
```

```
#pandasÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(df)Îäî ÏïÑÏ£ºÏïÑÏ£º Ï§ëÏöîÌï©ÎãàÎã§
print(df.head(0))
```

```
print(df.CONTENT)
```

```
corpus=''
for val in df.CONTENT:#valÏùò ÌÉÄÏûÖ : Serial! pandasÎ•º Íµ¨ÏÑ±ÌïòÎäî
  val = str(val)#Serial ->> strÌÉÄÏûÖÏúºÎ°ú Ìòï Î≥ÄÌôò
  tokens = val.split()#Í≥µÎ∞±Î¨∏ÏûêÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ï™ºÍ∞úÍ∏∞->listÌÉÄÏûÖÏúºÎ°ú Î≥ÄÌôò

  for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()#ÏÜåÎ¨∏ÏûêÎ°ú Î≥ÄÍ≤Ω
  #wordcloudÎ°ú ÌëúÌòÑÌï† ÎïåÏóêÎäî ÏÜåÎ¨∏Ïûê ÎåÄÎ¨∏ÏûêÎ•º Íµ¨Î≥ÑÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Î™®Îëê ÏÜåÎ¨∏ÏûêÎ°ú Î≥ÄÌôò
  for words in tokens:
    corpus = corpus + words + ' '#Îã®Ïñ¥Î•º Í≤∞Ìï©

print(corpus)
```

```
wordcloud = WordCloud().generate(corpus)

plt.figure(figsize = (8,8))#figure : Í∑∏Î¶ºÏùò ÏÇ¨Ïù¥Ï¶à
plt.imshow(wordcloud)#imshow : Ïù¥ÎØ∏ÏßÄÎ•º Î≥¥Ïó¨Ï§å
plt.axis("off")#Í∞ÄÎ°úÏ∂ï ÏÑ∏Î°úÏ∂ï ÎààÍ∏à x
plt.show()
```
